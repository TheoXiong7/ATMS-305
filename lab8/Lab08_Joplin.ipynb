```
# >>> F. OPEN FILE
# 1. Open the NetCDF file with Xarray's xr.open_dataset()
#      with:  DS = xr.open_dataset('name of the .nc file here');
# 2. Run just: DS
DS = xr.open_dataset('tornado12.nc')
DS
```

```
# >>> G. PLOT REFLECTIVITY (XY at ground)
# 1. Start a figure with figsize 10x8"
# 2. Set variable 'dbz' equal to: DS.Reflectivity.values
# 3. Make a color-filled contour plot like this:
#      plt.contourf(dbz[0,:],20,cmap='jet')
# 4. Add title: Simulated radar reflectivity at the ground
# 5. Add a colorbar with plt.colorbar()

fig1 = plt.figure(figsize=(10, 8))

dbz = DS.Reflectivity.values

plt.contourf(dbz[0,:],20,cmap='jet')
plt.title('Simulated radar reflectivity at the ground')
plt.colorbar()
```
