```python
# >> A: Import Numpy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# >> B: Vortex parameters (constants)
#   Max. tornado winds (m/s): set Vmax to 110
#   Radius of max winds (m) : set RMW = 75
Vmax = 110
RMW = 75
```

```python
# >> C: RADIUS ARRAY
# Create radius array running from 0 to 250 m, inclusive,
#   at 5 meter intervals.  Don't omit the last (250 m) value!
# Use the np.arange() function.
radius = np.arange(0, 251, 5)
radius
```

```python
# >> D: TANGENTIAL WIND ARRAY starting with ZEROES
# Create a Tangential velocity array full of zeros, named Vt.
# Make it the same size as your radius array, above.
# Use the np.zeros() function, and use radius.size with np.zeros.
Vt = np.zeros(len(radius))
Vt
```

```python
# >> E: LOOP, FILL IN "VT" ARRAY
# Use a for-loop to step through all values in your radius array.
#   The discusson below assumes your for-loop variable is "n".
# INSIDE the loop, set the tangential velocity Vt for each array element like so:
#   IF the radius radius[n] for *that element* is less than RMW:
#       set Vt[n] to Vmax*radius[n]/RMW
#   Otherwise -
#       set Vt[n] to Vmax*RMW/radius[n]
#
#   Finally, still in the loop, but OUTSIDE any if-else statements:
#     print out the radius and tangential wind for that element.
#   With the print inside a loop, you should get a list of values.
# *Format* your output like this:  (also to check your code)
#           For radius (m) =  40, Vt (m/s) =  58.67
#
#  (use 3 digits for radius -> whole numbers! - and format Vt as XXX.YY)
for n in range(len(radius)):
    if Vt[n] < RMW:
        Vt[n] = Vmax*radius[n]/RMW
    else:
        Vt[n] = Vmax*RMW/radius[n]
    print('For radius (m) = {:3}, Vt (m/s) = {:.2f}'.format(radius[n], Vt[n]))
```

```python
# >> F: PLOT TANGENTIAL WIND
# Plot the tangential wind (Y-axis) vs. radius (X-axis):
#  1. Make a figure that is 10x6"
#  2. Make a line plot of tangential wind Vt versus radius.
#       ... versus, means: radius is the x-axis, Vt on y.
#  3. Add x-label: Radius (meters)
#  4. Add y-label: Tangential Wind (m/s)
#  5. Add title: F: Tangential Wind vs. Radius
#  6. Add plt.grid() if you wish (try it!)
#  7. Finally, use plt.text() to put the word RMW at the peak
#       that is: in plt.text, the x-coordinate will be RMW
#                         and the y-coordinate will be Vmax.
#       ... include in the call to plt.text this:  horizontalalignment='center'
#       You should get letters RMW above peak of the curve.
fig = plt.figure(figsize=(10,6))

plt.plot(radius, Vt)
plt.xlabel('Radius (meters)')
plt.ylabel('Tangential Wind (m/s)')
plt.title('F: Tangential Wind vs. Radius')
plt.grid()
plt.text(RMW, Vmax, 'RMW', horizontalalignment='center')
```

```python
# >> G: BAR PLOT
# Bar plot: use plt.bar() as you did plt.plot() earlier.
# Add labels as before, on a 10x6" figure.
# Use title:  G: Tangential Wind vs. Radius
fig = plt.figure(figsize=(10,6))

plt.bar(radius, Vt)
plt.xlabel('Radius (meters)')
plt.ylabel('Tangential Wind (m/s)')
plt.title('G: Tangential Wind vs. Radius')
plt.grid()
plt.text(RMW, Vmax, 'RMW', horizontalalignment='center')
```

```python
# >> H: INTERROGATE your data
#  Let's pretend we just came across the radius() and vt() arrays;
#  ... perhaps we read them in (instead of creating them!)
#  1. Find the max value of array Vt; put answer into variable "maxwind"
#  2. Now let's use np.where() to find at what array **index** this max
#     value occurs.  Remember how to use np.where()!  You will need to put
#     [0][0] at the end of your np.where() statement.
#     Put the result into variable "index_Vtmax"
#
#  3. PRINT out the value of "maxwind" and the index_Vtmax value, in
#     a *formatted* print statement, so the results look like:
#        max wind = XXX.YY m/s, at array index XX
#
#  4. OK, now USE that index_Vtmax to print out the radius at [index_Vtmax]
#        as well as the value of Vt at [index_Vtmax] ... use
#        a *formatted* print statement so the results look like:
#             radius of max VT = XXX meters, where Vt = YYY m/s
maxwind = np.max(Vt)
index_Vtmax = np.where(Vt==maxwind)[0][0]
print('max wind = {:.2f} m/s, at array index {}'.format(maxwind, index_Vtmax))
```
