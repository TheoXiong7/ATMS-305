```python
# >> A. IMPORT
# Import what you need.  You'll be reading images and plotting images.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# >> B. RADAR and TRACK
# 1. Get this radar image using wget:  https://radar.weather.gov/ridge/standard/KLCH_0.gif
#       ... read it in with imread() into variable 'radar'.
# 2. Use the NHC Francine link on Canvas, go to Warnings/Cone Static Images and
#       right-click or otherwise get the URL of the forecast track image.
#       use wget to read it read it into variable 'track'.
# 3. Plot the radar and track data on a 1-row, 2-column figure.
#       Plot radar on left, and track on the right.  Add titles for each.
!wget -q -O KLCH_0.gif https://radar.weather.gov/ridge/standard/KLCH_0.gif
!wget -q -O track.png https://web.archive.org/web/20240911155012im_/https://www.nhc.noaa.gov/storm_graphics/AT06/refresh/AL062024_5day_cone_no_line_and_wind+png/152330_5day_cone_no_line_and_wind.png

radar = plt.imread('KLCH_0.gif')
track = plt.imread('track.png')

fig = plt.figure(dpi=200)

fig.add_subplot(1, 2, 1)
plt.imshow(radar)
plt.title('Hurricane Francine Radar')

fig.add_subplot(1, 2, 2)
plt.imshow(track)
plt.title('Hurricane Francine Track')

plt.tight_layout()
```

```python
# >> C: FLOODING
# 1. Again with the NHC Francine link on Canvas, go to the Flash Flooding Potential,
#       find the *image* URL, get with !wget and read into variable 'flood'.
# 2. Display the image with title: NHC Flash Flooding Potential

!wget -q -O flood.gif https://web.archive.org/web/20240911172108im_/https://www.nhc.noaa.gov/storm_graphics/AT06/refresh/AL0624WPCERO+gif/152330WPCERO_sm.gif

flood = plt.imread('flood.gif')

plt.imshow(flood)
plt.title('NHC Flash Flooding Potential')
```

```python
# >> D: SATELLITE and SURFACE
# 1. Go to NCAR/RAP Weather page
# 2. Select click/follow the options to get URLs for
#    a surface map and also a visible, infrared,
#    and water vapor satellite image -- on EACH, click on MGM
#    to select the image for the southeast U.S..
# 3. Get & read in all four images,
#      and display them in a 2x2 layout with subplot.
#    > Give each a title!

#!wget -q -O surface.gif https://weather.rap.ucar.edu/surface/2024092302_metars_cod.gif
#!wget -q -O visible.jpg https://weather.rap.ucar.edu/data/satellite/20240923/CONUS/color/20240923_021735_CONUS.jpg
#!wget -q -O ir.jpg https://weather.rap.ucar.edu/data/satellite/20240923/CONUS/C07/20240923_021736_CONUS.jpg
#!wget -q -O vapor.jpg https://weather.rap.ucar.edu/data/satellite/20240923/WMC/C09/20240923_021736_WMC.jpg

sur, vis, ir, vap = plt.imread('surface.gif'), plt.imread('visible.jpg'), plt.imread('ir.jpg'), plt.imread('vapor.jpg')

fig = plt.figure(figsize=(12,12), dpi=200)

fig.add_subplot(2, 2, 1)
plt.imshow(sur)
plt.title('Surface Map')

fig.add_subplot(2, 2, 2)
plt.imshow(vis)
plt.title('Satellite - Visible')

fig.add_subplot(2, 2, 3)
plt.imshow(ir)
plt.title('Satellite - Infrared')

fig.add_subplot(2, 2, 4)
plt.imshow(vap)
plt.title('Satellite - Vapor')

plt.tight_layout
```
